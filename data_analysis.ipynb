```python
kindle = pd.read_csv('/Users/aya/Desktop/amazon-reviews/datasets/kindle_reviews.csv')
```

```python
kindle.head()
```

```python
kindle = kindle[['asin', 'reviewerID', 'reviewTime', 'reviewerName', 'summary', 'reviewText', 'overall']]
new_column_names = {
    'asin': 'asin',
    'overall': 'reviews.rating',
    'reviewText': 'reviews.text',
    'reviewTime': 'reviewTime',
    'reviewerID': 'reviewerID',
    'reviewerName': 'reviewerName',
    'summary': 'reviews.title'
}

# Rename the columns using the new names
kindle = kindle.rename(columns=new_column_names)
kindle.head()
```

```python
# Specify the desired subset size
subset_size = 6000

# Group the dataset by the specified column
grouped = kindle.groupby('reviews.rating')

# Calculate the number of data points to sample from each group
sample_per_group = subset_size // len(grouped)

# Take a random subset from each group
subset_df = grouped.apply(lambda x: x.sample(n=sample_per_group, replace=False)).reset_index(drop=True)

subset_df.head()

```

```python
# Save the subset DataFrame to a CSV file
subset_df.to_csv('kindle_reviews_6000.csv', index=False)
```

```python
# Specify the desired subset size
subset_size = 10000

# Group the dataset by the specified column
grouped = kindle.groupby('reviews.rating')

# Determine the minimum number of samples per group
min_samples_per_group = subset_size * 0.01  # 1% of total

# Take the minimum sample from each group, or all data points if a group is smaller
subset_df = pd.concat([
    group.sample(n=min(int(min_samples_per_group), len(group)), replace=False) 
    for _, group in grouped
]).reset_index(drop=True)

# If we have less than subset_size data points, sample more from the entire dataset
if len(subset_df) < subset_size:
    additional_samples = subset_size - len(subset_df)
    remaining_df = df.loc[~df.index.isin(subset_df.index)]
    # If there are fewer remaining data points than additional_samples, take them all
    if len(remaining_df) <= additional_samples:
        subset_df = pd.concat([subset_df, remaining_df])
    else:
        remaining_samples = remaining_df.sample(n=additional_samples, replace=False)
        subset_df = pd.concat([subset_df, remaining_samples])

# If we have more than subset_size data points, remove some randomly
elif len(subset_df) > subset_size:
    subset_df = subset_df.sample(n=subset_size, replace=False)
```
